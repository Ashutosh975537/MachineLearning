```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler 
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.drop(columns='Survived'),df[['Survived']],test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
df.isnull().sum()
```

```python
si = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')
```

```python
x_train_age = si.fit_transform(x_train[['Age']])
```

```python
x_train_age
```

```python
x_train_embark = si_embarked.fit_transform(x_train[['Embarked']])
```

```python
x_train_embark[2]
```

```python
x_test_age = si.transform(x_test[['Age']])
x_test_age.shape
```

```python
x_test_embarked = si_embarked.transform(x_test[['Embarked']])
```

```python
x_test_embarked[1]
```

```python
ohe = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
x_train_gender = ohe.fit_transform(x_train[['Sex']])
x_train_embark = ohe_embarked.fit_transform(x_train_embark)
```

```python
x_test_gender = ohe.transform(x_test[['Sex']])
```

```python
x_train_gender
```

```python
x_test_embarked = ohe_embarked.transform(x_test[['Embarked']])
```

```python
x_train_embark
```

```python
x_train_rem = x_train.drop(columns=['Sex','Age','Embarked'])
x_test_rem = x_test.drop(columns=['Sex','Age','Embarked'])
```

```python
x_test_rem
```

```python
x_train_transformed = np.concatenate((x_train_rem,x_train_age,x_train_gender,x_train_embark),axis=1)
```

```python
x_train_transformed
```

```python
x_test_transformed = np.concatenate((x_test_rem,x_test_age,x_test_gender,x_test_embarked),axis = 1)
```

```python
x_test_transformed
```

```python
type(y_train)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transformed,y_train)
```

```python
y_pred = clf.predict(x_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred)
```
