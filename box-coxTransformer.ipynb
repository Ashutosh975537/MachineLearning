```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
x = df.drop(columns=['Strength'])
```

```python
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=
                                                       0.2,random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

# Applying the transformation

```python
for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x[col], dist="norm", plot=plt)
    plt.title(col)

    plt.show()
```

```python
pt = PowerTransformer(method='box-cox')
```

```python
x_train_transformed = pt.fit_transform(x_train+0.000001)
x_test_tranformed = pt.transform(x_test+0.000001)
```

```python
x_train_transformed
```

```python
pt.lambdas_
```

```python
pd.DataFrame({'cols' : x_train.columns,'box-cox-lamdas' : pt.lambdas_})
```

```python
x_train.describe()
```

```python
lr = LinearRegression()
lr.fit(x_train_transformed,y_train)
```

```python
y_pred = lr.predict(x_test_tranformed)
```

```python
r2_score(y_test,y_pred)
```

```python
# Applying the cross validation here
pt = PowerTransformer(method='box-cox')
X_transformed = pt.fit_transform(x+0.0000001)

lr = LinearRegression()
np.mean(cross_val_score(lr,X_transformed,y,scoring='r2'))
```
